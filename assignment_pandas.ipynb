```python
import pandas as pd
import numpy as np
```

```python
# загружаем датасет
data = pd.read_csv("./data/adult.data.csv")
data.head()
```

**1. Посчитайте, сколько мужчин и женщин (признак *sex*) представлено в этом датасете**

```python
display(data.groupby(by=['sex'], as_index=False) \
            .agg(count=('sex','count')))
```

**2. Каков средний возраст мужчин (признак *age*) по всему датасету?**

```python
float(round(data[data['sex'] == 'Male']['age'].mean(), 2))
```

**3. Какова доля граждан Соединенных Штатов (признак *native-country*)?**

```python
round((data[data['native-country'] == 'United-States'].shape[0]) / (data.shape[0]), 2) * 100
```

**4-5. Рассчитайте среднее значение и среднеквадратичное отклонение возраста тех, кто получает более 50K в год (признак *salary*) и тех, кто получает менее 50K в год**

```python
display(data.groupby(by=['salary'], as_index=False) \
            .agg(count=('salary','count')))
```

```python
data.groupby(by=['salary'], as_index=False) \
    .agg(mean_age=('age', 'mean'), std_age=('age', 'mean'))
```

**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
data[data['salary'] == '>50K'].groupby(by=['marital-status']).count()
```

```python
data['married'] = data['marital-status'].apply(lambda x: 'married' if 'Married' in x else 'not married')
married_percent = data[data['salary'] == '>50K'].groupby(by=['married']).agg(percent=('married', 'count')).apply(lambda x: x / sum(x) * 100)
married_percent[married_percent['percent'] == married_percent['percent'].max()]
```

**9. Какое максимальное число часов человек работает в неделю (признак *hours-per-week*)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
max_hours_per_week = int(data['hours-per-week'].max())
max_hours_per_week
```

```python
people_max_time = int(data[data['hours-per-week'] == max_hours_per_week]['hours-per-week'].count())
people_max_time
```

```python
percent_high_salary_max_time = float(round(data[(data['hours-per-week'] == max_hours_per_week) & (data['salary'] == '>50K')]['salary'].count() \
                               / people_max_time * 100, 2))
percent_high_salary_max_time
```

**10. Посчитайте среднее время работы (*hours-per-week*) зарабатывающих мало и много (*salary*) для каждой страны (*native-country*).**

```python
data.groupby(by=['salary', 'native-country'], as_index=False).agg(avg_hours=('hours-per-week', 'mean'))
```

**11.Сгруппируйте людей по возрастным группам *young*, *adult*, *retiree*, где:**
* *young* соответствует 16-35 лет

```python
data['AgeGroup'] = data['age'].apply(lambda x: 'young' if x < 35 else ('adult' if x >= 35 & x < 70 else 'retiree'))
data
```

**12-13. Определите количество зарабатывающих >50K в каждой из возрастных групп (колонка AgeGroup), а также выведите название возрастной группы, в которой чаще зарабатывают больше 50К (>50K)**

```python
age_high_salary = data[data['salary'] == '>50K'].groupby(by=['AgeGroup'], as_index=False) \
                              .agg(count=('AgeGroup', 'count')) 
age_high_salary
```
